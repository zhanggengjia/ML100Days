```python
import matplotlib.pyplot as plt
import numpy as np

#one quarter cos
x = np.arange(0, 180)
y = np.cos(x*np.pi / 180.0)
plt.plot(x, y)
plt.show()

#one and half sin
x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
plt.plot(x, y_sin)
plt.show()

#one cos
x = np.arange(0, 3*np.pi, 0.1)
y_cos = np.cos(x)
plt.plot(x, y_cos)
plt.show()
```

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

dot = np.arctan2(Y, X)

plt.scatter(X, Y, s = 65, c = dot, alpha = 0.5)


print(dot)
```
